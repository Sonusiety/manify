In this notebook, we explore whether it's possible to create a decision tree that's identical to the one we define
simply by composing preprocessing and postprocessing steps:
1. Preprocess 1: convert coordinates to angles
2. Preprocess 2: convert circular coordinates to binary features
3. Train sklearn on specially-preprocessed features
4. Postprocess: replace midpoint angles with recomputed midpoint angles

```python
import embedders
```

```python
pm = embedders.manifolds.ProductManifold(signature=[(-1,2), (0, 2), (1, 2)])
X, y = embedders.gaussian_mixture.gaussian_mixture(pm=pm)
```

```python
pdt = embedders.tree_new.ProductSpaceDT(pm=pm)
angles, labels, classes, comparisons = pdt._preprocess(X, y)
```
