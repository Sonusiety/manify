```python
import manify
```

```python
pm = manify.manifolds.ProductManifold(signature=[(1, 2), (1, 2), (0, 2), (-1, 2), (-1, 2)])

X, y = pm.gaussian_mixture(1000)
```

```python
# Sketch of how this would work
import torch

X_factorized = pm.factorize(X)

subtree_indices = torch.randint(0, len(pm.P), (10,))

trees = []
for idx in subtree_indices:
    pm_sub = manify.manifolds.ProductManifold([pm.signature[idx]])
    X_sub = X_factorized[idx]
    tree = manify.predictors.tree_icml.ProductSpaceDT(pm=pm_sub)
    tree.fit(X_sub, y)
    trees.append(tree)

[tree.predict(X_factorized[idx]) for idx in subtree_indices]
```

```python
from manify.predictors.tree_icml import SingleManifoldEnsembleRF

rf = SingleManifoldEnsembleRF(pm=pm, n_estimators=10)
rf.fit(X, y)
rf.predict(X)
rf.score(X, y)
```
